```python
x = ['Age','Annual_Income','Num_of_Loan', 'Num_of_Delayed_Payment', 'Outstanding_Debt', 'Amount_invested_monthly', 'Monthly_Balance']
for i in x:
  df[f'{i}']=pd.to_numeric(df[f'{i}'],errors='coerce')
```

```python
df.Credit_Mix = df.Credit_Mix.replace({"_":np.nan})
```

```python
df = df.fillna(method='ffill')
df = df.fillna(method='bfill')
```

```python
df.Month = df.Month.map({'January': 0, 'February': 1, 'March': 3, 'April': 4,'May':5,'June':6,'July': 7,'August':8})
```

```python
df = df.iloc[:,4:]
```

```python
for i in range(len(df.Credit_History_Age.values)):
  df.Credit_History_Age[i] = int(df.Credit_History_Age.values[i].split('Years and ')[0])*12+ int(df.Credit_History_Age.values[i].split('Years and ')[1].replace(' Months',''))
```

```python
df['Credit_History_Age'] = pd.to_numeric(df['Credit_History_Age'])
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],train_size=0.7,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred_rfc = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred_rfc)
```

```python
print(confusion_matrix(y_test,y_pred_rfc))
```
